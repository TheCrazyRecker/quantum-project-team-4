```python
import math
def ncz_qi(n, rot, had):
    if n == 2:
        if had:
            print("H q[2]")
        print(f"CR q[1], q[2], {rot}")
        print("CNOT q[0], q[1]")
        print("CNOT q[1], q[0]")
        print(f"CR q[1], q[2], {-rot}")
        print("CNOT q[0], q[1]")
        print(f"CR q[1], q[2], {rot}")
        if had:
            print("H q[2]")
    else:
        if had:
            print(f"H q[{n}]")
        print(f"CR q[{n-1}], q[{n}], {rot}")
        ncz_qi(n - 1, 1.5708, True)
        print(f"CR q[{n-1}], q[{n}], {-rot}")
        ncz_qi(n - 1, 1.5708, True)
        print(f"SWAP q[{n-1}], q[{n}]")
        ncz_qi(n - 1, rot / 2, False)
        print(f"SWAP q[{n-1}], q[{n}]")
        if had:
            print(f"H q[{n}]")
     
def set_element_qi(element: int, n: int):
    binary_string =  bin(element).removeprefix('0b')
    while len(binary_string) < n:
        binary_string = '0' + binary_string
    if '0' not in binary_string:
        return
    binary_string = reversed(binary_string)
    res = '{'
    for idx, binary in enumerate(binary_string):
        if binary == str(0):
            res += f' X q[{idx}] |'
    res = res[:len(res) - 2]
    res += ' }'
    print(res)

def grover_qi(n: int, element: int):
    if element >= math.pow(2, n):
        print('Element does not fall into the range [0, 2^n - 1]')
        return
    print('version 1.0 \n')
    print(f'qubits {n} \n')
    print(".init")
    print(f"H q[0:{n-1}] \n")
    print(f'.grover({math.floor((math.pi / 4) * math.sqrt(math.pow(2, n)))})')
    print("#.oracle")
    set_element_qi(element, n)
    ncz_qi(n - 1, 1.5708, False)
    print("SWAP q[0], q[1] \n")
    print("#.diffusion")
    set_element_qi(element, n)
    print(f"H q[0:{n - 1}]")
    print(f"X q[0:{n - 1}]")
    ncz_qi(n - 1, 1.5708, False)
    print("SWAP q[0], q[1]")
    print(f"X q[0:{n - 1}]")
    print(f"H q[0:{n - 1}]")
    
grover_qi(4, 5)
```
